```python
a = input("a = ")
b = input("b = ")

if a > b:
    print("a > b")
else:
    print("a <= b")
    
if a > b:
    print("a > b")
elif a == b:
    print("a = b")
else:
    print("a < b")
```

```python
result = 0
i = 1
n = 10
while i <= n:
    print("i = ", i)
    result += i
    i += 1

print("result = ", result)
```

```python
result = 0
i = 1
n = 100
while i <= n:
    print("i = ", i)
    result += i
    i += 2

print("result = ", result)
```

```python
result = 0
for i in range(0, -100, -1):
    if i % 2 != 0:
        result += i
    print("i = ", i)
    
print("result = ", result)
```
